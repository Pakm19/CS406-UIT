```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
forest = '/content/drive/MyDrive/Baitap/Baitap1/images/forest.tif'
img = cv2.imread(forest)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
a = img.copy()
image = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
new_image = np.zeros(image.shape, image.dtype)
alpha = 1.0 # Simple contrast control
beta = 0    # Simple brightness control
# Initialize values
print(' Basic Linear Transforms ')
print('-------------------------')
try:
    beta = int(input('* Enter the beta value [-100 -> 100]: '))
except ValueError:
    print('Error, not a number')

for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            new_image[y,x,c] = np.clip(alpha*image[y,x,c] + beta, 0, 255)

# Show stuff
cv2_imshow(image)
cv2_imshow(new_image)
```

```python
b = img.copy()
image = cv2.cvtColor(b, cv2.COLOR_BGR2RGB)
new_image = np.zeros(image.shape, image.dtype)
alpha = 1.0 # Simple contrast control
beta = 0    # Simple brightness control
# Initialize values
print(' Basic Linear Transforms ')
print('-------------------------')
try:
  alpha = float(input('* Enter the alpha value [1.0-3.0]: '))
except ValueError:
    print('Error, not a number')

for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            new_image[y,x,c] = np.clip(alpha*image[y,x,c] + beta, 0, 255)

# Show stuff
cv2_imshow(image)
cv2_imshow(new_image)
```

Yêu cầu 4: Invert cho ảnh:Kid.tif

```python
def invert(imagem):
    imagem = (255-imagem)
    return imagem
```

```python
def show2image(img_src, img_rst, title1, title2):
  plt.figure(figsize = (20, 20))
  plt.subplot(1, 2, 1)
  plt.title(title1)
  img_src = cv2.cvtColor(img_src, cv2.COLOR_BGR2RGB)
  img_src = plt.imshow(img_src, interpolation = 'bicubic')
  plt.subplot(1, 2, 2)
  plt.title(title2)
  img_rst = cv2.cvtColor(img_rst, cv2.COLOR_BGR2RGB)
  img_rst = plt.imshow(img_rst, interpolation = 'bicubic')
```

```python
baby = cv2.imread("/content/drive/MyDrive/Baitap/Baitap1/images/kids.tif")

kids_invert = invert(baby)
show2image(baby, kids_invert, 'Anh goc', 'inverted')
```
